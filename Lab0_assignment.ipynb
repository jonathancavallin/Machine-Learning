```python
import matplotlib.pyplot as plt
import csv
from scipy import stats
import numpy as np
import sklearn as sl
from sklearn import linear_model
```

```python
# Load the provided data file with the used car data (you can also have a look at it with any text editor)


# place your loading code here
filename = "km_year_power_price (2).csv"
lines = csv.reader(open(filename, newline=''), delimiter=',')
km = []
price= []
power= []
year= []


#print(data)
lines= list(lines)
lines.remove(lines[0])


for i, line in enumerate(lines):
     km.append(float(line[0]))
     year.append(float(line[1]))
     power.append(float(line[2]))
     price.append(float(line[3]))

#print(km)
km = np.array(km)
year = np.array(year)
power = np.array(power)
price = np.array(price)
    
```

Use linear regression to estimate the car prices from the year, kilometers or engine power. 
You can make a simple 1D regression from each one of the parameters independently 




```python
# linear regression with linregress (estimate price from year)
# your code.....

price_year= stats.linregress(year, price)

estimated_price_year= price_year.slope*year+price_year.intercept
print(estimated_price_year)

print("The R^2 value is " + str(price_year.rvalue**2))
```

```python
# Plot the data and the lines representing the output of the linregress and least square algorithms

# your code....
import matplotlib.pyplot as plt

plt.plot(year,price,'o',label='real data with year')
plt.plot(year, estimated_price_year, 'r', label= 'linear fit with year')
plt.xlabel('year')
plt.ylabel('price ($)')
plt.legend()
plt.show()
```

```python
# linear regression with linregress (estimate price from power)

# your code.....

price_power= stats.linregress(power, price)

estimated_price_power= price_power.slope*power+price_power.intercept
print(estimated_price_power)

print("The R^2 value is " + str(price_power.rvalue**2))
plt.plot(power,price,'o',label='real data with power')
plt.plot(power, estimated_price_power, 'r', label= 'linear fit with power')
plt.xlabel('power (PS) ')
plt.ylabel('price ($)')
plt.legend()
plt.show()
```

```python
# linear regression with linregress (estimate price from km)

# your code...
price_km= stats.linregress(km, price)

estimated_price_km= price_km.slope*km+price_km.intercept
print(estimated_price_km)
print("The R^2 value is " + str(price_km.rvalue**2))

plt.plot(km,price,'o',label='real data with kilometres')
plt.plot(km, estimated_price_km, 'r', label= 'linear fit with kilometers')
plt.xlabel('km')
plt.ylabel('price ($)')
plt.legend()
plt.show()
```

```python
# Have a look at the correlation coefficients to see which of the 3 features works better

# your code......

print("The R^2 value for the year fit is" + str(price_year.rvalue**2))
print("The R^2 value for the power fit is " + str(price_power.rvalue**2))
print("The R^2 value for the km fit is " + str(price_km.rvalue**2))
```

The best parameter to predict the price of the car according to these data is its power since the correlation of the variables is the highest.
